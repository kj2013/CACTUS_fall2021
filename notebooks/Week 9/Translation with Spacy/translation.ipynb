```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("Helsinki-NLP/opus-mt-zh-en",model_max_length  = 511)

model = AutoModelForSeq2SeqLM.from_pretrained("Helsinki-NLP/opus-mt-zh-en")
```

```python
from tqdm import tqdm
tqdm.pandas()
```

```python
import pandas as pd
import torch
```

```python
device = "cuda:4" if torch.cuda.is_available() else "cpu"
```

```python
device
```

```python
model = model.to(device)
```

```python
df = pd.read_csv('Zbcomma.csv')
```

```python
df = df[df['message'].notna()]
```

```python
df
```

```python
df.message.iloc[1]
```

```python
def translate(sample_text):
    print(sample_text)
    batch = tokenizer([sample_text], return_tensors="pt" , truncation=True).to(device)
    gen = model.generate(**batch).to(device)
    torch.cuda.empty_cache()
    return tokenizer.batch_decode(gen, skip_special_tokens=True)[0]
```

```python
df['translated_message'] = df["message"].progress_apply(lambda x : translate(x))
```

```python
df.to_csv('Zbcomma_translated.csv' , index = False)
```
